**Q1: Run K-means clustering with Euclidean, Cosine and Jarcard similarity. Specify K= the number of categorical values of y (the number of classifications). Compare the SSEs of Euclidean-K-means, Cosine-K-means,Jarcard-K-means. Which method is better?  (10 points)**

```python
# Number of clusters based on labels
k = len(np.unique(labels))
print(f'Count of unique labels in label.csv aka number of clusters is {k}')
```

```python
# Distance metrics calculations
def euclidean_distance(x, y):
    return np.linalg.norm(x - y)

def cosine_distance(x, y):
    return 1 - np.dot(x, y) / (np.linalg.norm(x) * np.linalg.norm(y))

def jaccard_distance(x, y):
    numerator = np.minimum(x, y).sum()
    denominator = np.maximum(x, y).sum()
    return 1 - (numerator / denominator) if denominator != 0 else 1
```

```python
# Calculate Sum of Squared Errors
def compute_sse(X, clusters, centroids):
    sse = 0
    for cluster_idx, cluster in enumerate(clusters):
        for point_idx in cluster:
            sse += euclidean_distance(X[point_idx], centroids[cluster_idx]) ** 2
    return sse
```

```python
# Custom K-Means algorithm
def kmeans_clustering(X, k, distance_fn, max_iters=500, tol=1e-4):
    np.random.seed(42)
    centroids = X[np.random.choice(X.shape[0], k, replace=False)]  # Initialize centroids

    for iteration in range(max_iters):
        # Step 1: Assign points to the nearest centroid
        clusters = [[] for _ in range(k)]
        for idx, point in enumerate(X):
            distances = [distance_fn(point, centroid) for centroid in centroids]
            cluster_idx = np.argmin(distances)
            clusters[cluster_idx].append(idx)

        # Step 2: Update centroids
        new_centroids = []
        for cluster in clusters:
            if cluster:  # Avoid empty clusters
                new_centroids.append(np.mean(X[cluster], axis=0))
            else:
                new_centroids.append(centroids[np.random.choice(range(k))])  # Reinitialize empty clusters
        new_centroids = np.array(new_centroids)

        # Step 3: Check for convergence
        centroid_shift = np.sum([euclidean_distance(old, new) for old, new in zip(centroids, new_centroids)])
        if centroid_shift < tol:
            break
        centroids = new_centroids

    return clusters, centroids
```

```python
def run_experiment(X, labels, k):
    results = {}
    distance_metrics = {
        "Euclidean Distance": euclidean_distance,
        "Cosine Similarity": cosine_distance,
        "Jaccard Similarity": jaccard_distance,
    }

    for name, distance_fn in distance_metrics.items():
        clusters, centroids = kmeans_clustering(X, k, distance_fn)
        sse = compute_sse(X, clusters, centroids)
        results[name] = {"SSE": sse}

    return results
```

```python
# Run the experiment
results = run_experiment(data, labels, k)

# Print results
for metric, stats in results.items():
    print(f"{metric} - SSE: {stats['SSE']:.2f}")
```

From the above results, we can infer that **Euclidean** Distance is the best method as it provides the lowest SSE value.

**Q2: Compare the accuracies of Euclidean-K-means Cosine-K-means, Jarcard-K-means. First, label each cluster using the majority vote label of the data points in that cluster. Later, compute the predictive accuracy of Euclidean-K-means, Cosine-K-means, Jarcard-K-means. Which metric is better? (10 points)**

```python
# Step 1 - Label each cluster using majority vote
def majority_vote(labels, clusters):
    cluster_labels = []
    for cluster in clusters:
        if cluster:
            cluster_label = Counter(labels[cluster]).most_common(1)[0][0]
        else:
            cluster_label = -1  # Handle empty clusters
        cluster_labels.append(cluster_label)
    return cluster_labels
```

```python
# Step 2 - Compute accuracy of each metric
def compute_accuracy(labels, clusters, cluster_labels):
    correct = 0
    total = 0
    for cluster_idx, cluster in enumerate(clusters):
        for point_idx in cluster:
            if labels[point_idx] == cluster_labels[cluster_idx]:
                correct += 1
            total += 1
    return correct / total
```

```python
def run_experiment(X, labels, k):
    results = {}
    distance_metrics = {
        "Euclidean": euclidean_distance,
        "Cosine": cosine_distance,
        "Jaccard": jaccard_distance,
    }

    for name, distance_fn in distance_metrics.items():
        clusters, centroids = kmeans_clustering(X, k, distance_fn)
        cluster_labels = majority_vote(labels, clusters)
        accuracy = compute_accuracy(labels, clusters, cluster_labels)
        results[name] = {"Accuracy": accuracy}

    return results
```

```python
results = run_experiment(data, labels, k)

# Print results
for metric, stats in results.items():
    print(f"{metric} - Accuracy: {stats['Accuracy']:.2%}")
```

**Q3: Set up the same stop criteria: when there is no change in centroid position OR when the SSE value increases in the next iteration OR when the maximum preset value (e.g., 500, you can set the preset value by yourself) of iteration is complete,  for Euclidean-K-means, Cosine-K means, Jarcard-K-means. Which method requires more iterations and times to converge? (10 points)**

```python
# Modified K-means with new stopping criteria
def k_means_with_criteria(data, k, distance_function, max_iters=100):
    n_samples = data.shape[0]
    centroids = data[np.random.choice(n_samples, k, replace=False)]
    prev_centroids = np.zeros_like(centroids)
    labels = np.zeros(n_samples, dtype=int)
    prev_sse = float("inf")
    iteration = 0

    start_time = time.time()

    for iteration in range(1, max_iters + 1):
        # Step 1: Assign labels based on closest centroid
        distances = np.array([[distance_function(sample, centroid) for centroid in centroids] for sample in data])
        labels = np.argmin(distances, axis=1)

        # Step 2: Recompute centroids
        new_centroids = np.array([data[labels == j].mean(axis=0) if np.any(labels == j) else centroids[j] for j in range(k)])

        # Step 3: Compute SSE
        sse = sum(np.sum((data[labels == j] - centroids[j]) ** 2) for j in range(k))

        # Step 4: Check stopping criteria
        if np.allclose(new_centroids, centroids, atol=1e-6):
            print(f"Converged: No change in centroids at iteration {iteration}")
            break
        if sse > prev_sse:
            print(f"Stopped: SSE increased at iteration {iteration}")
            break

        centroids = new_centroids
        prev_sse = sse

    end_time = time.time()

    # Return results
    return {
        "clusters": [data[labels == j] for j in range(k)],
        "centroids": centroids,
        "sse": sse,
        "iterations": iteration,
        "time": end_time - start_time
    }

# Run K-means for each distance metric and compare iterations and time
results = {}
for metric_name, distance_func in {
    'Euclidean': euclidean_distance,
    'Cosine': cosine_distance,
    'Jaccard': jaccard_distance,
}.items():
    print(f"Running K-means with {metric_name} distance...")
    result = k_means_with_criteria(data, k, distance_func)
    results[metric_name] = result
    print(f"{metric_name}: Iterations = {result['iterations']}, Time = {result['time']:.2f} seconds")

# Summary of results
print("\nSummary of Results:")
for metric_name, result in results.items():
    print(f"{metric_name}: Iterations = {result['iterations']}, Time = {result['time']:.2f} seconds")
```

**Euclidean K-Means** requires the most iterations and the most time to converge.

```python
# Modified K-means to support multiple stopping conditions
def k_means_multiple_stops(data, k, distance_function, max_iters=100):
    n_samples, n_features = data.shape
    centroids = data[np.random.choice(n_samples, k, replace=False)]
    prev_centroids = np.zeros_like(centroids)
    labels = np.zeros(n_samples, dtype=int)
    prev_sse = float("inf")
    sse_stopping_conditions = {}

    start_time = time.time()

    for iteration in range(1, max_iters + 1):
        # Step 1: Assign labels based on closest centroid
        distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)

        # Step 2: Recompute centroids
        new_centroids = np.array([
            data[labels == j].mean(axis=0) if np.any(labels == j) else centroids[j]
            for j in range(k)
        ])

        # Step 3: Compute SSE
        sse = sum(np.sum((data[labels == j] - centroids[j]) ** 2) for j in range(k))

        # Step 4: Check each stopping condition
        if "centroid_no_change" not in sse_stopping_conditions and np.allclose(new_centroids, centroids, atol=1e-6):
            sse_stopping_conditions["centroid_no_change"] = (iteration, sse)

        if "sse_increase" not in sse_stopping_conditions and sse > prev_sse:
            sse_stopping_conditions["sse_increase"] = (iteration, prev_sse)
            break  # SSE increase ends the run

        centroids = new_centroids
        prev_sse = sse

    # Stopping due to max iterations
    if "max_iterations" not in sse_stopping_conditions:
        sse_stopping_conditions["max_iterations"] = (iteration, sse)

    end_time = time.time()

    return {
        "sse_conditions": sse_stopping_conditions,
        "time": end_time - start_time
    }

# Run K-means for each distance metric and stopping condition
results = {}
for metric_name, distance_func in {
    'Euclidean': euclidean_distance,
    'Cosine': cosine_distance,
    'Jaccard': jaccard_distance,
}.items():
    print(f"Running K-means with {metric_name} distance...")
    result = k_means_multiple_stops(data, k, distance_func)
    results[metric_name] = result
    print(f"{metric_name}: {result['sse_conditions']}")

# Output SSE comparisons
print("\nSSE Comparisons for Different Stopping Conditions:")
for metric_name, result in results.items():
    print(f"\n{metric_name}:")
    print(f"  Time taken: {result['time']:.2f} seconds")
    for condition, (iterations, sse) in result["sse_conditions"].items():
        print(f"  {condition}: Iterations = {iterations}, SSE = {sse:.2f}")
```
